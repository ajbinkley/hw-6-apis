```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from pprint import pprint

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)


# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

## Perform API Calls

```python

city_data = []

# Starting URL for Weather Map API Call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key

for city in cities:
    request_url = requests.get(url + '&q=' + city)
    city_json = request_url.json()
    city_data.append(city_json)

pprint(city_data)    
```

```python
city_names, cloudiness, country, date, humidity, lat, lng, max_temp, wind_speed = ([] for i in range(9))

for i in range(len(city_data)):
    
    try:
        city_names.append(city_data[i]['name'])
    except KeyError:
        city_names.append('Not found')
        
    try:
        cloudiness.append(city_data[i]['clouds']['all'])
    except KeyError:    
        cloudiness.append('Not found')
        
    try:    
        country.append(city_data[i]['sys']['country'])
    except KeyError:
        country.append('Not found')
         
    try:
        date.append(city_data[i]['dt'])
    except KeyError:
        date.append('Not found')
        
    try:
        humidity.append(city_data[i]['main']['humidity'])
    except KeyError:
        humidity.append('Not found')
        
    try:
        lat.append(city_data[i]['coord']['lat'])
    except KeyError:
        lat.append('Not found')
        
    try:
        lng.append(city_data[i]['coord']['lon'])
    except KeyError:
        lng.append('Not found')
        
    try:
        max_temp.append(city_data[i]['main']['temp_max'])
    except KeyError:
        max_temp.append('Not found')
        
    try:
        wind_speed.append(city_data[i]['wind']['speed'])
    except KeyError:
        wind_speed.append('Not found')
```

```python
city_data_df = pd.DataFrame({
    'City' : city_names,
    'Cloudiness' : cloudiness,
    'Country' : country,
    'Date' : date,
    'Humidity' : humidity,
    'Lat' : lat,
    'Lng' : lng,
    'Max Temp' : max_temp,
    'Wind Speed' : wind_speed 
})

city_data_df.head(20)
```

```python
city_data_df = city_data_df.loc[city_data_df['City'] != 'Not found']
city_data_df.head(20)
```

```python
x_lat = city_data_df['Lat']
y_temp = city_data_df['Max Temp']

plt.scatter(x_lat, y_temp)
plt.xlabel('Latitude')
plt.ylabel('Max Temperature ('F')')
plt.suptitle('City Latitude vs Max Temperature (10/2/2018)')
```

```python
x_lat = city_data_df['Lat']
y_humid = city_data_df['Humidity']

plt.scatter(x_lat, y_humid)
plt.xlabel('Latitude')
plt.ylabel('Humidity (%)')
plt.suptitle('City Latitude vs Humidity (10/2/2018)')
```

```python
x_lat = city_data_df['Lat']
y_cloudy = city_data_df['Cloudiness']

plt.scatter(x_lat, y_cloudy)
plt.xlabel('Latitude')
plt.ylabel('Cloudiness (%)')
plt.suptitle('City Latitude vs Cloudiness (10/2/2018)')
```

```python
x_lat = city_data_df['Lat']
y_wind = city_data_df['Wind Speed']

plt.scatter(x_lat, y_wind)
plt.xlabel('Latitude')
plt.ylabel('Wind Speed (mph)')
plt.suptitle('City Latitude vs Wind Speed (10/2/2018)')
```
